**Q2. Find the Pivot Integer**

<br>

- Given a positive integer *n*, find the pivot integer *x* such that:

- The sum of all elements between **1** and **x** inclusively equals the sum of all elements between **x** and **n** inclusively.
- Return the pivot integer *x*. If no such integer exists, return *-1*. It is guaranteed that there will be at most one pivot index for the given input.

<p><strong class="example">Example 1:</strong></p>
<pre><strong>Input:</strong> n = 8
<strong>Output:</strong> 6
<strong>Explanation:</strong> 6 is the pivot integer since: 1 + 2 + 3 + 4 + 5 + 6 = 6 + 7 + 8 = 21.
</pre>

<p><strong class="example">Example 2:</strong></p>
<pre><strong>Input:</strong> n = 1
<strong>Output:</strong> 1
<strong>Explanation:</strong> 1 is the pivot integer since: 1 = 1.
</pre>

<p><strong class="example">Example 3:</strong></p>
<pre><strong>Input:</strong> n = 4
<strong>Output:</strong> -1
<strong>Explanation:</strong> It can be proved that no such integer exist.
</pre>

<br>
<p><strong>Constraints:</strong></p>
<li><code>1 &lt;= n &lt;= 1000</code></li>
<br><br>

<p><strong class="example">Solution :</strong></p>

```python
class Solution(object):
    def pivotInteger(self, n):
        ans = -1
        arr = [ i+1 for i in range(n)]
        for i in range(n):
            if sum(arr[:i+1]) == sum(arr[i:]):
                ans = arr[i]
                break
        return ans
```

```python
import random

obj = Solution()

for i in range(10):
    n = random.randint(1, 1000)
    print(n," : ",obj.pivotInteger(n))
```
