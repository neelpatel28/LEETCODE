```python
class Solution(object):
    def getBiggestThree(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: List[int]
        """
        m, n = len(grid), len(grid[0])
        s = set()
        for i in range(m):
            for j in range(n):
                s.add(grid[i][j])
                for k in range(1, min(i, j, m - 1 - i, n - 1 - j) + 1):
                    s.add(grid[i - k][j] + grid[i + k][j] + grid[i][j - k] + grid[i][j + k] + grid[i - k][j - k] + grid[i - k][j + k] + grid[i + k][j - k] + grid[i + k][j + k])
        return sorted(s, reverse=True)[:3]
```

```python
class Solution(object):
    def getBiggestThree(self, grid):
        m, n = len(grid), len(grid[0])
        def rhombus_sum(r, c, k):
            if k == 0:
                return grid[r][c]
            coords = set()
            for d in range(k + 1):
                coords.add((r - k + d, c + d))
                coords.add((r + d, c + k - d))
                coords.add((r + k - d, c - d))
                coords.add((r - d, c - k + d))
            return sum(grid[x][y] for x, y in coords)

        s = set()
        for i in range(m):
            for j in range(n):
                s.add(grid[i][j])
                max_k = min(i, j, m - 1 - i, n - 1 - j)
                for k in range(1, max_k + 1):
                    s.add(rhombus_sum(i, j, k))
        return sorted(s, reverse=True)[:3]

sol = Solution()
print(sol.getBiggestThree(grid))
print(sol.getBiggestThree(grid1))
```
